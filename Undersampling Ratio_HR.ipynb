```python
df_dt=df.iloc[:5,:]
df_RF=df.iloc[5:10,:]
df_LR=df.iloc[10:15,:]
df_SVM=df.iloc[15:20,:]
df_NB=df.iloc[20:25,:]
df_GB=df.iloc[25:30,:]
df_KNN=df.iloc[30:35,:]
```

```python
def plot_score(data,k):
    plt.figure()
    plt.plot( 'Ratio', 'AUC', data=data, marker='o', markerfacecolor='palegreen', markersize=12, color='darkred', linewidth=4)
    plt.plot( 'Ratio', 'Recall', data=data, marker='v', markerfacecolor='darkturquoise', markersize=12, color='darkred', linewidth=4)
    plt.plot( 'Ratio', 'Accuracy', data=data, marker='8', markerfacecolor='royalblue', markersize=12, color='darkred', linewidth=4)
    plt.legend()
    plt.title(k)
    plt.xlabel("Ratio")
    plt.ylabel("Score")
```

```python
for i in range(1):
    plot_score(df_dt,k="DT")
    plt.savefig('DT_Undersampling_1')
    plot_score(df_RF,k="RF")
    plt.savefig('RF_Undersampling_1')
    plot_score(df_LR,k="LR")
    plt.savefig('LR_Undersampling_1')
    plot_score(df_SVM,k="SVM")
    plt.savefig('SVM_Undersampling_1')
    plot_score(df_NB,k="NB")
    plt.savefig('NB_Undersampling_1')
    plot_score(df_GB,k="GB")
    plt.savefig('GB_Undersampling_1')
    plot_score(df_KNN,k="KNN")
```
